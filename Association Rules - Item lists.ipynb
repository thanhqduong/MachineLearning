```python
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0,
                                                        df.shape[1])])
```

```python
records[0]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df['nan']
```

```python
df = df.drop(columns=['nan'])
```

```python
df.columns.tolist()
```

```python
frequent_itemsets = apriori(df, min_support=0.003, use_colnames=True)
frequent_itemsets.head()
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.003)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.003)
rules
```

```python
for row in rules.iterrows():
    if "sausage" in row[1][0]:
        print(row)
```

```python
support=rules['support'].values
confidence=rules['confidence'].values
lift = rules['lift'].values
```

```python
plt.figure(figsize=(8,5))
plt.scatter(support, confidence, s= lift*100,alpha=0.8, c = lift)
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
